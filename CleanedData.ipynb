```python

import pandas as pd
import matplotlib
import requests
import io
import numpy as np
url = "https://raw.githubusercontent.com/mich-matocha/data-therapy/main/data/QA_2016.T.csv"
url_content = requests.get(url).text

#with open()
data = pd.read_csv(url,encoding='utf-8')

data.head()
```

```python
data = data.replace('-1',np.NaN) #?
data.head()
```

```python
index = 0
for row in data['3']:
    if row != 'United States of America':
        data.drop(index=index,inplace=True)
    index += 1
data.head()
```

```python
print('size',data.shape)
print('index',data.index)

data_index = data.index

index = 0 
for row in data['5']:
    if row == 1:
        data.drop(index=data_index[index],inplace=True)
    index += 1
print('new size',data.shape)
data.head()
```
