```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df_2016 = pd.read_csv('https://raw.githubusercontent.com/mich-matocha/data-therapy/main/data/QA_2016_clean.csv', encoding='utf-8')
```

```python
# Do you feel that your employer takes mental health as seriously as physical health?

plt.pie(df_2016[['UserID', '91']].groupby('91').count()['UserID'].values, labels=['I don\'t know', 'No', 'Yes'])
plt.title('Do you feel that your employer takes mental health as seriously as physical health?')

plt.show()
```

```python
df16_2 = df_2016.dropna(subset=['2'])

df16_m = df16_2.loc[df16_2['2'] == 'Male']
df16_f = df16_2.loc[df16_2['2'] == 'Female']
df16_q = df16_2[~df16_2['2'].str.contains('Male|Female')]

fig, axs = plt.subplots(1, 3)
axs[0].pie(df16_m[['UserID', '91']].groupby('91').count()['UserID'].values, labels=['I don\'t know', 'No', 'Yes'])
axs[1].pie(df16_f[['UserID', '91']].groupby('91').count()['UserID'].values, labels=['I don\'t know', 'No', 'Yes'])
axs[2].pie(df16_q[['UserID', '91']].groupby('91').count()['UserID'].values, labels=['I don\'t know', 'No', 'Yes'])
plt.title('Do you feel that your employer takes mental health as seriously as physical health?')
plt.show()
```
